## 🔍 Key Finding: Natural Control Group Exists!

**This is excellent news for our analysis.**

- **42.4% of students (340/801)** received **0 NCRBS! lessons** 
- The remaining **57.6%** received some level of intervention (1-20 lessons)

This means we have a natural **"control group"** and **"treatment group"** to compare! 

**However, the critical question remains**: 
- Were students with 0 lessons simply not selected for intervention (perhaps because they were already performing well)?
- Or was intervention offered universally and some students just didn't complete lessons?

**Next step**: Let's examine whether baseline PM1 scores differ between the intervention and non-intervention groups. This will help us understand selection bias.

---

```python
# =============================================================================
# SELECTION BIAS CHECK: Do baseline scores differ by intervention group?
# =============================================================================

# Create intervention groups
df_complete['Intervention_Group'] = pd.cut(
    df_complete['NCRBS_Lessons'], 
    bins=[-1, 0, 10, 20], 
    labels=['No Intervention (0)', 'Low Dose (1-10)', 'High Dose (11-20)']
)

# Compare baseline PM1 scores across groups
print("=" * 70)
print("BASELINE (PM1) SCORES BY INTERVENTION GROUP")
print("=" * 70)

baseline_comparison = df_complete.groupby('Intervention_Group').agg({
    'FAST PM1 Score %': ['count', 'mean', 'std', 'median'],
    'Pretest Uppercase Letter Names (out of 26)': 'mean',
    'Pretest Lowercase Letter Names (out of 26)': 'mean',
    'Pretest Letter Sounds (out of 26)': 'mean'
}).round(3)

print(baseline_comparison)

# Statistical test: Are baseline scores significantly different?
from scipy import stats

no_intervention = df_complete[df_complete['NCRBS_Lessons'] == 0]['FAST PM1 Score %'].dropna()
with_intervention = df_complete[df_complete['NCRBS_Lessons'] > 0]['FAST PM1 Score %'].dropna()

t_stat, p_value = stats.ttest_ind(no_intervention, with_intervention)
print(f"\n--- T-test: No Intervention vs. Any Intervention ---")
print(f"No Intervention Mean PM1: {no_intervention.mean():.3f}")
print(f"With Intervention Mean PM1: {with_intervention.mean():.3f}")
print(f"T-statistic: {t_stat:.3f}, P-value: {p_value:.4f}")
```

## 🚨 CRITICAL FINDING: Selection Bias Confirmed

**The data shows STRONG evidence that the intervention was targeted at struggling students:**

| Group | PM1 Baseline Score | Letter Knowledge (avg) |
|-------|-------------------|----------------------|
| No Intervention | **59.4%** | 16.5 uppercase, 13.8 lowercase, 9.7 sounds |
| Low Dose (1-10 lessons) | **27.1%** | 8.6 uppercase, 7.4 lowercase, 4.0 sounds |
| High Dose (11-20 lessons) | **27.7%** | 7.5 uppercase, 6.3 lowercase, 4.5 sounds |

**Statistical significance:** T-statistic = 19.87, p < 0.0001

### What this means:
1. ✅ **Good news**: The district appears to have correctly targeted the intervention to students who needed it most
2. ⚠️ **Analytical challenge**: We **cannot** simply compare outcomes between intervention/no-intervention groups because they started at very different baselines
3. 🎯 **Next step**: We need to look at **change in scores** (improvement) rather than raw end-of-year scores

### Recommended approach:
- Calculate PM3 - PM1 (improvement) for each student
- Compare improvement rates between groups
- Consider: Did low-performing students with intervention "catch up" to their peers?
